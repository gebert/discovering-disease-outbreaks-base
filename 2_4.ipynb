```python
import pandas as pd
from unidecode import unidecode
import numpy as np

df = pd.read_pickle("./data/dataframe1.pkl")
```

```python
alternative_names = {unidecode(name['name']): name['name'] for name in gc.get_cities().values()}
                     
#print(gc.get_cities_by_name(alternative_names['Hagatna']))
```

```python
longs = []
lats = []
coco = []

#count = 0

df = df.fillna("")

for x in df.iterrows():
    city = x[1]["cities"]

    if city != "":
        # City name available
        # Get all cities by name
        city_list = gc.get_cities_by_name(alternative_names[city])
        anz = len(city_list)
        if anz > 0:
            # There is a result
            # Find the city with the highest population even though this approach might not be the best. A different approach might be to analyse where the majority of the cases are from and to select the city with the smalest distance
            best_city = max(city_list,
                          key=lambda x: list(x.values())[0]['population'])
            citydata = list(best_city.values())[0]

#            if citydata["geonameid"] != list(gc.get_cities_by_name(alternative_names[city])[0].values())[0]["geonameid"]:
#                count = count+1
                
            longs.append(citydata["longitude"])
            lats.append(citydata["latitude"])
            coco.append(citydata["countrycode"])
            #print(f"Es wurden {anz} verschiedene {x} gefunden. Long: {long}")
        else:
            print(f"No entry for {city} found.")
    else:
        # No city given
        longs.append("")
        lats.append("")
        coco.append("")

assert len(longs) == len(lats) == len(coco)

#print(f"Longs: {len(longs)} Cities: {len(df)}")
#print(f"Städte mit mehr als einem Ergebnis: {count}")
#print(longs)
```

```python
df["latitude"] = lats
df["longitude"] = longs
df["countrycode"] = coco
df = df.drop("countries","columns")
df = df[df.latitude != ""]
```
